# Opdracht Titanic
### Load libraries

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

### Load en bekijk data

```python
data = pd.read_csv("titanic_data.csv")
data.head()
```

### Bekijk welke data categorisch is

```python
data.select_dtypes(exclude=[np.number]).columns
```

### Verijder de kolommen 'Name', 'Cabin' en 'Ticket'

```python
# PassengerId is ook onnodig
data = data.drop(["Name", "Cabin", "Ticket", "PassengerId", "SibSp"], axis=1)

data
```

### Codeer categorische features met indices (getallen)

```python
# Multiple categorical columns
categorical_cols = ['Sex', 'Embarked']

data = pd.get_dummies(data, columns=categorical_cols)
data
```

### Vervang NaN in "Ages". Bepaal zelf of mediaan of gemiddelde hiervoor het best is.

```python
gemiddelde = round(data["Age"].mean())

data['Age'] = data['Age'].fillna(gemiddelde)

data
```

### Bereid de data voor om je model te trainen. Wat wordt je target vector en hoe ziet je feature matrix er uit?

```python
# Target is Survived en the feature matrix is eigenlijk de vorige dataset (zonder name, tickets etc.) minusde survived column 

X = data.drop('Survived', axis=1)
print(X.shape)
y = data['Survived']
print(y.shape)
# split to train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=6)
```

### Train en test KNN classifier

```python
# choose a model
knn = KNeighborsClassifier(n_neighbors=12)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)

knn_acc = accuracy_score(y_test, y_pred)
print(knn_acc)
```

### Train en test regressie classifier. Kies je lineair of logistisch?

```python
# Logistisch, want de survivability is een binaire regressie, oftwel ja/nee antwoord

lr = LogisticRegression(max_iter=700)
lr.fit(X_train, y_train)

y_pred = lr.predict(X_test)

lr_acc = accuracy_score(y_pred, y_test)

lr_acc

```

### Welk model geeft het beste resultaat?

```python
# Logistisch geeft betere resultaat
print('knn accuracy: ', knn_acc)
print('logistisch accuracy: ', lr_acc)
```
