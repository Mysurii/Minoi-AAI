```python
import pandas as pd
```

## Data inladen

```python
# Data die ontbreekt: Instrument bij Theon
# Opschonen is nodig, zo zie je bij studierichting van tyrion HBO-ICT, wat niet daar hoort of Theon's leeftijd die 119 is
# Opschonen kan m.b.v. python pandas library

data1 = pd.read_csv('ijsbrekervragen.csv')
data2 = pd.read_csv('ijsbrekervragen2.csv')

frames = [data1, data2]
df = pd.concat(frames, ignore_index=True)

df
```

```python
# verander de ja/nee naar booleans 1/0
df['Heb je goede wifi thuis?'] = df['Heb je goede wifi thuis?'].map({'ja': 1, 'nee': 0})
df['Heb je huisdieren?'] = df['Heb je huisdieren?'].map({'ja': 1, 'nee': 0})


df.dtypes
```

```python
import numpy as np
# fix HBO_ICT studierichting to SE (meest voorkomend)

df['Studierichting'].replace('HBO-ICT', 'SE', inplace=True)

df
```

```python
# Fix instrument empty cell
df['Welke instrument speel je?'].fillna('geen',inplace=True)
df
```

```python
# Fix leeftijd van 119 naar 19
df.loc[5, 'Leeftijd'] = '19'
df
```

```python
# Fix foute data -> Hond naar 'ja'
df.loc[6, 'Heb je huisdieren?'] = 1
df
```

```python
count = df[df['Studierichting'] == 'SE']['Studierichting'].count()
total = df.shape[0]

se_percentage = count / total * 100

se_percentage
```

```python
mean = df['Reistijd in minuten'].mean()
median = df['Reistijd in minuten'].median()
minimum = df['Reistijd in minuten'].min()
maximum = df['Reistijd in minuten'].max()

print('mean', mean)
print('median', median)
print('min', minimum)
print('max', maximum)
```

```python
col = df[(df['Heb je goede wifi thuis?'] == False) & (df['Heb je huisdieren?'] == False)]['Heb je goede wifi thuis?'].count()
total = df['Heb je goede wifi thuis?'].count()

chance = col / total * 100

chance
```

```python
column_geen = df[df['Welke instrument speel je?'] == 'geen']['Reistijd in minuten']
column_wel = df[df['Welke instrument speel je?'] != 'geen']['Reistijd in minuten']

minimum_reistijd_geen = column_geen.min()
gem_reistijd_geen = column_geen.mean()
maximum_reistijd_geen = column_geen.max()

minimum_reistijd_wel = column_wel.min()
gem_reistijd_wel = column_wel.mean()
maximum_reistijd_wel = column_wel.max()


print('min geen:', minimum_reistijd_geen)
print('gem geen:', gem_reistijd_geen)
print('max geen:', maximum_reistijd_geen)

print('\nmin wel:', minimum_reistijd_wel)
print('gem wel:', gem_reistijd_wel)
print('max wel:', maximum_reistijd_wel)
```

```python
# one hot encoding
pd.get_dummies(df.Studierichting, prefix='Studierichting')
```
